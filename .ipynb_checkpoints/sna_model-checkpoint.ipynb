Model 2.2 Proximal only repression

```python
def sna_prox_repression(lvl, t,
                       c, b_prox, b_dist,
                       K_ploc, K_pfar,
                       K_X, n_X,
                       n_prox, y_sna,
                       b_SNA, y_SNA,
                       b_x, y_x, K_SNAx, n_SNAx,
                       b_X, y_X):
    
    sna, SNA, x, X = lvl
    
    dsna_dt = (c * (K_X**n_X  / (K_X**n_X + X**n_X)) * ((b_prox * (K_ploc**n_prox / (K_ploc**n_prox + SNA**n_prox)))
               + (b_dist * (K_pfar**n_prox / (K_pfar**n_prox + SNA**n_prox))))
               - y_sna * sna)
    dSNA_dt = b_SNA * sna - y_SNA * SNA
    
    dx_dt = b_x * (K_SNAx**n_SNAx  / (K_SNAx**n_SNAx + SNA**n_SNAx)) - y_x * x
    dX_dt = b_X * x - y_X * X
    
    return (dsna_dt, dSNA_dt, dx_dt, dX_dt)
```

Model 2.0 Proximal and Distal repression

```python
def sna_all_repression(lvl, t,
                       c, b_prox, b_dist,
                       K_ploc, K_pfar, K_dloc, K_dfar,
                       K_X, n_X,
                       n_prox, n_dist, y_sna,
                       b_SNA, y_SNA,
                       b_x, y_x, K_SNAx, n_SNAx,
                       b_X, y_X):
    
    sna, SNA, x, X = lvl
    
    dsna_dt = (c * (K_X**n_X  / (K_X**n_X + X**n_X))
               * ((b_prox * (K_ploc**n_prox / (K_ploc**n_prox + SNA**n_prox)) * (K_dfar**n_dist / (K_dfar**n_dist + SNA**n_dist)))
               + (b_dist * (K_dloc**n_dist / (K_dloc**n_dist + SNA**n_dist)) * (K_pfar**n_prox / (K_pfar**n_prox + SNA**n_prox))))
               - y_sna * sna)
    dSNA_dt = b_SNA * sna - y_SNA * SNA
    
    dx_dt = b_x * (K_SNAx**n_SNAx  / (K_SNAx**n_SNAx + SNA**n_SNAx)) - y_x * x
    dX_dt = b_X * x - y_X * X
    
    return (dsna_dt, dSNA_dt, dx_dt, dX_dt)
```

```python
def sna_local_repression(lvl, t,
                       c, b_prox, b_dist,
                       K_ploc, K_dloc,
                       K_X, n_X,
                       n_prox, n_dist, y_sna,
                       b_SNA, y_SNA,
                       b_x, y_x, K_SNAx, n_SNAx,
                       b_X, y_X):
    
    sna, SNA, x, X = lvl
    
    dsna_dt = (c * (K_X**n_X  / (K_X**n_X + X**n_X)) * ((b_prox * (K_ploc**n_prox / (K_ploc**n_prox + SNA**n_prox)))
               + (b_dist * (K_dloc**n_dist / (K_dloc**n_dist + SNA**n_dist))))
               - y_sna * sna)
    dSNA_dt = b_SNA * sna - y_SNA * SNA
    
    dx_dt = b_x * (K_SNAx**n_SNAx  / (K_SNAx**n_SNAx + SNA**n_SNAx)) - y_x * x
    dX_dt = b_X * x - y_X * X
    
    return (dsna_dt, dSNA_dt, dx_dt, dX_dt)
```

```python
def solve_beta(r, n):
    numerator = (2 * r**(n+1)) - 1
    denominator = r**(n+1) - r
    m = numerator/denominator
    m = m.astype(dtype=np.complex128)

    if m > 0:
        B = ((2 - m)/(m**(n+1)))**(1/n)
    else:
        B =np.nan

    if np.isreal(B):
        B = np.real(B)
    else:
        B = np.nan
    
    return B
```

```python
def solve_n_m(R, r_m, r_c, n_c):
    m_c_n = (2 * r_c**(n_c+1)) - 1
    m_c_d = r_c**(n_c+1) - r_c
    m_c = m_c_n/m_c_d
    #m = m.astype(dtype=np.complex128)
    
    log_n = 1 - (R * r_m * m_c)
    log_d = 2 - (R * m_c)

    if (log_d != 0) and (log_n/log_d > 0):
        n_m = (np.log(log_n/log_d)/np.log(r_m)) - 1
    else:
        n_m = np.nan

    return n_m
```

```python
arr_ratio_nond = arr_1copy / arr_2copy
arr_ratio_nond[-2] = 0.99
#arr = np.sort(arr_ratio_nond)
#sorted_indices = np.argsort(arr_ratio_nond)

n_c = np.linspace(0.01, 100, 10000)
B_c = np.zeros([len(arr_ratio_nond), len(n_c)])
n_m = np.zeros([len(arr_ratio_nond), len(n_c)])
B_m = np.zeros([len(arr_ratio_nond), len(n_c)])

for j, n_ in enumerate(n_c):
    for i, r in enumerate(arr_ratio_nond):
        B_c[i,j] = solve_beta(r, n_)
        n_m[i,j] = solve_n_m(arr_2copy[i], r, arr_ratio_nond[0], n_)


for i, r in enumerate(arr_ratio_nond):
    for j, n_ in enumerate(n_m[i,:]):
        B_m[i,j] = solve_beta(r, n_)
```

```python
def sna_simple_nond(lvl, t, n, B, y, c):
    
    m, p = lvl

    dm_dt = (c/(1+p**n)) - m

    dp_dt = (B * m - p) * y
    
    return (dm_dt, dp_dt)
```

```python
def sna1_simple_nond(lvl, t, B, y, c):
    
    m, p = lvl

    dm_dt = c - m

    dp_dt = (B * m - p) * y
    
    return (dm_dt, dp_dt)
```

```python
def hill_fun(m, y, n):
    return 1 / ((y*m)**n + 1)
```

Model 2.0 with X

```python
def root_fun_2_0(m, A, D, E, F, G, H, I, J, n, n_p, n_d, c):
    X = hill_fun(hill_fun(m, D, n), A, n)
    pp = hill_fun(m, E, n_p)
    pd = hill_fun(m, F, n_d)
    dd = hill_fun(m, G, n_d)
    dp = hill_fun(m, H, n_p)
    return c * X * ((I * pp * pd) + (J * dd * dp))- m

def _msolve_2_0(A, D, E, F, G, H, I, J, n, n_p, n_d, c):
    return scipy.optimize.brentq(root_fun_2_0, 0, 10, args=(A, D, E, F, G, H, I, J, n, n_p, n_d, c))
```

```python
def fit_model_2_0(x, A, D, E, F, G, H, I, J, n, n_p, n_d,
                     H_13, n_p_13,
                     E_3, H_3, I_3, n_p_3,
                     E_26, H_26, I_26, n_p_26,
                     E_26_dl, H_26_dl, I_26_dl, n_p_26_dl,
                     E_dl, H_dl, I_dl, n_p_dl,
                     F_04, G_04, J_04, n_d_04):

    y = np.zeros(len(x))
    
    for i in range(len(x)):
        #if (x[i] > 0) and (x[i] < 21):
        #    yw_2 = _msolve_2_0(A, D, E, F, G, H, I, J, n, n_p, n_d, 2)
        #else:
        #    y[i] = np.nan
        #    continue
     
        # yw 2 copy
        if x[i] >= 0.5 and x[i] < 1.5:
            y[i] = _msolve_2_0(A, D, E, F, G, H, I, J, n, n_p, n_d, 2)
            #y[i] = yw_2/yw_2
    
        # yw 1 copy
        elif x[i] >= 1.5 and x[i] < 2.5:
            y[i] = _msolve_2_0(A, D, E, F, G, H, I, J, n, n_p, n_d, 1)#/yw_2
    
        # sna1 2 copy
        elif x[i] >= 2.5 and x[i] < 3.5:
            y[i] = _msolve_2_0(A, 0, 0, 0, 0, 0, I, J, n, 1, 1, 2)#/yw_2
    
        # sna1 1 copy
        elif x[i] >= 3.5 and x[i] < 4.5:
            y[i] = _msolve_2_0(A, 0, 0, 0, 0, 0, I, J, n, 1, 1, 1)#/yw_2
    
        # 1.3 2 copy
        elif x[i] >= 4.5 and x[i] < 5.5:
            y[i] = _msolve_2_0(A, D, 0, 0, G, H_13, 0, J, n, n_p_13, n_d, 2)#/yw_2
    
        # 1.3 1 copy
        elif x[i] >= 5.5 and x[i] < 6.5:
            y[i] = _msolve_2_0(A, D, 0, 0, G, H_13, 0, J, n, n_p_13, n_d, 1)#/yw_2
    
        # 3.0 2 copy
        elif x[i] >= 6.5 and x[i] < 7.5:
            y[i] = _msolve_2_0(A, D, E_3, F, G, H_3, I_3, J, n, n_p_3, n_d, 2)#/yw_2
    
        # 3.0 1 copy
        elif x[i] >= 7.5 and x[i] < 8.5:
            y[i] = _msolve_2_0(A, D, E_3, F, G, H_3, I_3, J, n, n_p_3, n_d, 1)#/yw_2
    
        # 4.4 2 copy
        elif x[i] >= 8.5 and x[i] < 9.5:
            y[i] = _msolve_2_0(A, D, 0, 0, G, 0, 0, J, n, 1, n_d, 2)#/yw_2
    
        # 4.4 1 copy
        elif x[i] >= 9.5 and x[i] < 10.5:
            y[i] = _msolve_2_0(A, D, 0, 0, G, 0, 0, J, n, 1, n_d, 1)#/yw_2
    
        # 2.6 2 copy
        elif x[i] >= 10.5 and x[i] < 11.5:
            y[i] = _msolve_2_0(A, D, E_26, F, G, H_26, I_26, J, n, n_p_26, n_d, 2)#/yw_2
    
        # 2.6 1 copy
        elif x[i] >= 11.5 and x[i] < 12.5:
            y[i] = _msolve_2_0(A, D, E_26, F, G, H_26, I_26, J, n, n_p_26, n_d, 1)#/yw_2
    
        # 2.6 dl l>h 2 copy
        elif x[i] >= 12.5 and x[i] < 13.5:
            y[i] = _msolve_2_0(A, D, E_26_dl, F, G, H_26_dl, I_26_dl, J, n, n_p_26_dl, n_d, 2)#/yw_2
    
        # 2.6 dl l>h 1 copy
        elif x[i] >= 13.5 and x[i] < 14.5:
            y[i] = _msolve_2_0(A, D, E_26_dl, F, G, H_26_dl, I_26_dl, J, n, n_p_26_dl, n_d, 1)#/yw_2
    
        # dl l>h 2 copy
        elif x[i] >= 14.5 and x[i] < 15.5:
            y[i] = _msolve_2_0(A, D, E_dl, F, G, H_dl, I_dl, J, n, n_p_dl, n_d, 2)#/yw_2
    
        # dl l>h 1 copy
        elif x[i] >= 15.5 and x[i] < 16.5:
            y[i] = _msolve_2_0(A, D, E_dl, F, G, H_dl, I_dl, J, n, n_p_dl, n_d, 1)#/yw_2
    
        # 0.4 2 copy
        elif x[i] >= 16.5 and x[i] < 17.5:
            y[i] = _msolve_2_0(A, D, E, F_04, G_04, H, I, J_04, n, n_p, n_d_04, 2)#/yw_2
    
        # 0.4 1 copy
        elif x[i] >= 17.5 and x[i] < 18.5:
            y[i] = _msolve_2_0(A, D, E, F_04, G_04, H, I, J_04, n, n_p, n_d_04, 1)#/yw_2
    
        # 2.0 2 copy
        elif x[i] >= 18.5 and x[i] < 19.5:
            y[i] = _msolve_2_0(A, D, E, 0, 0, 0, I, 0, n, n_p, 1, 2)#/yw_2
    
        # 2.0 1 copy
        elif x[i] >= 19.5 and x[i] < 20.5:
            y[i] = _msolve_2_0(A, D, E, 0, 0, 0, I, 0, n, n_p, 1, 1)#/yw_2
            
        else:
            y[i] = np.nan

    return y
```
